is produced:

Output:

=========================

```python
class DemonSlayer:
  count = 0
  
  def __init__(self, name, age, style, kills):
    self.name = name
    self.age = age
    self.style = style
    self.kills = kills
    DemonSlayer.count += 1
  
  def addKill(self, k):
    self.kills += k
  
  def dismissKill(self, k):
    if k < self.kills:
      self.kills -= k
  
  def printDetails(self):
    print(f"Name: {self.name}\nAge:{self.age}\nStyle:{self.style}\nTotal Kill:{self.kills}")

print('No of demon slayers:', DemonSlayer.count)
print("=========================")
d1 = DemonSlayer("Tanjiro", 15, "Water Breathing", 100)
d1.addKill(20)
d1.printDetails()
print("=========================")
d2 = DemonSlayer("Rengoku", 20, "Flame Breathing", 1000)
d2.dismissKill(50)
d2.printDetails()
print("=========================")
d3 = DemonSlayer("Zenitsu", 16, "Thunder Breathing", 80)
d3.dismissKill(100)
d3.printDetails()
print("=========================")
print('No of demon slayers: ', DemonSlayer.count)
```
