```python
my_tuple = (("Firmino", 1101),("Gabriel", 1201),("Casemiro", 1302),("Fred", 1402),("Messi", 1111),("Aguero", 1211),("Klose", 1121), ("Werner", 1221), ("Toni", 1322))
country = {(2, 0): "Brazil", (2, 1): "Argentina", (2, 2): "Gremany", (3, 1): "Forward", (3, 2): "Midfield"}
output = {}

for x in my_tuple:
    Last_digit = int(str(x[1])[3])
    Third_digit = int(str(x[1])[2])

    if country[(2, Third_digit)] in output:
        if country[(3, Last_digit)] in output[country[(2, Third_digit)]]:
            lst = list(output[country[(2, Third_digit)]][country[(3, Last_digit)]])
            lst.append(x[0])
            output[country[(2, Third_digit)]][country[(3, Last_digit)]] = tuple(lst)

        else:
            output[country[(2, Third_digit)]][country[(3, Last_digit)]] = (x[0],)

    else:
        dict = {country[(3, Last_digit)]: (x[0],)}
        output[country[(2, Third_digit)]] = dict
        
print(output, "\n")
```
