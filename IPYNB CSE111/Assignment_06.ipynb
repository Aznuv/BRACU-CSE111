```python
#Task 2
class Assassin:
    NumAssassin = 0
    def __init__(self, name, rate):
        self.name = name 
        self.rate = rate
        Assassin.NumAssassin += 1
        
    def printDetails(self):
        print(f"Name: {self.name} \nSuccess rate: {self.rate}% \nTotal number of Assassin: {Assassin.NumAssassin}")
    
    @classmethod
    def failureRate(cls, name, perc):
        rate = 100 - perc
        obj = cls(name, rate)
        return obj
    
    @classmethod
    def failurePercentage(cls, name, perc):
        rate = 100 - perc
        obj = cls(name, rate)
        return obj
              
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage('Akabane', 10)
akabane.printDetails()
```

```python
#Task 05
class Employee:
    def __init__(self, name, peri):
        self.name = name
        self.workingPeriod = peri
    
    @classmethod
    def employeeByJoiningYear(cls, name, time):
        peri = 2021 - time
        obj= cls(name, peri)
        return obj
    @staticmethod
    def experienceCheck(exp, gen):
        if exp < 3:
            ex = "not experienced"
        else:
            ex = "experienced"
            
        if gen == "male": 
            gend = "He"
        else: 
            gend = "She"
        return f"{gend} is {ex}"
        
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task 06
class Laptop:
    laptopCount = 0
    def __init__(self, name, count):
        self.name = name
        self.count = count
        Laptop.laptopCount += count
    @staticmethod
    def advantage():
        print("Laptops are portable")
    @classmethod
    def resetCount(cls):
        cls.laptopCount = 0
lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
# Task 07
class Cat:
    Number_of_cats = 0
    def __init__(s, cr = "white", position = "sitting"):
        s.cr = cr
        s.position = position
        Cat.Number_of_cats = Cat.Number_of_cats + 1

    @classmethod
    def no_parameter(cls):
        cr = "White"
        position = "sitting"
        obj = cls(cr,position)
        return obj

    @classmethod
    def first_parameter(cls, cr):
        cls.cr = cr
        position = "sitting"
        obj = cls(cr, position)
        return obj

    @classmethod
    def second_parameter(cls, position):
        cr = "White"
        cls.position = position
        obj = cls(cr, position)
        return obj

    def changeColor(s, cr):
        s.cr = cr

    def printCat(s):
        print(f"{s.cr} cat is {s.position}")

print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task 08
import math
class Cylinder:
    radius = 5
    height = 18

    def __init__(s, radius, height):
        s.radius = radius
        s.height = height
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}.")
        Cylinder.radius = radius
        Cylinder.height = height
        print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}.")

    @classmethod
    def swap(cls, radius, height):
        obj = cls(height,radius)
        return obj

    @classmethod
    def changeFormat(cls, all):
        new_r,new_h = all.split("-")
        new_obj = cls (float(new_r),float(new_h))
        return new_obj

    @staticmethod
    def volume(radius, height):
        print("Volume:",(math.pi)*radius*radius*height)

    @staticmethod
    def area(radius, height):
        print("Area:",(2*math.pi*radius*radius)+(2*math.pi*radius*height))

c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task 09
class Student:
    allStu = 0
    bracStu = 0
    othStu = 0
    
    def __init__(self, name, dept, ins = None):
        self.name = name
        self.dept = dept
        if ins == None:
            self.ins = "BRAC University"
            Student.bracStu += 1
        else: 
            self.ins = ins
            Student.othStu +=1
        Student.allStu += 1
            
    @classmethod
    def createStudent(cls, name, dept, ins = None):
        if ins == None:
            obj = cls(name, dept, "BRAC University")
        else:
            obj = cls(name, dept, ins)
        return obj

    def individualDetail(self):
        print("Name:", self.name)
        print("Department:", self.dept)
        print("Institution:", self.ins)
        
    @classmethod
    def printDetails(cls):
        print("Total Student(s):", cls.allStu)
        print("BRAC University Student(s):", cls.bracStu) 
        print("Other Institution Student(s):", cls.othStu)
        
Student.printDetails()
print('#########################')
mikasa = Student('Mikasa Ackerman', "CSE")
mikasa.individualDetail()
print('------------------------------------------')
Student.printDetails()
print('========================')
harry = Student.createStudent('Harry Potter', "Defence Against Dark Arts", "Hogwarts School")
harry.individualDetail()
print('-------------------------------------------')
Student.printDetails()
print('=========================')
levi = Student.createStudent("Levi Ackerman", "CSE")
levi.individualDetail()
print('--------------------------------------------')
Student.printDetails()
```

```python
#Task 10
class SultansDine:
    tot_branch = 0
    tot_sell = 0
    list1 = []
    
    def __init__(self, name):
        self.name = name
        self.branch_sell = 0
        SultansDine.tot_branch += 1
        SultansDine.list1.append(self)
        
    def sellQuantity(self, quantity):
        self.quantity = quantity
        
        if self.quantity < 10:
            self.branch_sell = self.quantity * 300
        elif self.quantity < 20:
            self.branch_sell = self.quantity * 350
        else:
            self.branch_sell = self.quantity * 400
        
        SultansDine.tot_sell += self.branch_sell
            
    def branchInformation(self):
        print('Branch Name:', self.name)
        print('Branch Sell:', self.branch_sell, 'Taka')
        
    @classmethod
    def details(cls):
        print('Total Number of branch(s):', SultansDine.tot_branch)
        print('Total Sell:', SultansDine.tot_sell, 'Taka')
        for obj in SultansDine.list1:
            print('Branch Name:', obj.name, end = ', ')
            print('Branch Sell:', obj.branch_sell)
            percentage = (obj.branch_sell / SultansDine.tot_sell) * 100
            print("Branch consists of total sell's: {}%".format("%.2f" % percentage))
        
SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
