```python
#Task 6
class Vehicle:
    def __init__ (self, x = 0, y = 0):
        self.x = x
        self.y = y
    def print_position(self):
        print("(", self.x, ",", self.y, ")", sep = "")
    def moveUp (self):
        self.y = self.y + 1
    def moveDown(self):
        self.y = self.y - 1
    def moveLeft (self):
        self.x = self.x - 1
    def moveRight (self):
        self.x = self.x + 1

car = Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#Task 9
class Patient:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def add_Symptom(self, *info):
        self.symptom = info 

    def printPatientDetail(self):
        print(f"Name:{self.name}\nAge:{self.age}")
        print("Symtoms:", end="")
        for i in range(len(self.symptom)):
            if i < (len(self.symptom) - 1):
                print(self.symptom[i], end=',')
            else:
                print(self.symptom[i])
    
p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")

print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
print("=========================")
```

```python
#Task 15
class Account:
    def __init__(self, name = None, cash = None ):
        if name == None:
            self.name = "Default Account"
        else:
            self.name = name 
        if cash == None:
            self.balance = 0.0
        else:
            self.balance = cash

    def withdraw(self, cash2):
        self.cash2 = cash2
        
        if self.cash2 <= self.balance:
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
        else:
            z = self.cash2 - self.balance
            self.balance = z
            print("Withdraw successful! New balance is:", z) 
    
    def details(self):
        print(self.name)
        print(self.balance)
        
a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930);
print("------------------------")
a2.withdraw(600);
print("------------------------")
a1.withdraw(6929)
```
